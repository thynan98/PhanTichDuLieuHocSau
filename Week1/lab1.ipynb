```python
#Cau 6 
import pandas as pd
import numpy as np

df = pd.read_csv("dulieuxettuyendaihoc.csv")

score_cols = [col for col in df.columns 
              if col.startswith(('T','L','H','S','V','X','D','N','DH'))
              and col not in ['DT']]

for col in score_cols:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
#Cau 7
def tinh_TBM(df, suffix):
    return (df[f'T{suffix}']*2 + df[f'L{suffix}'] + df[f'H{suffix}'] +
            df[f'S{suffix}'] + df[f'V{suffix}']*2 + df[f'X{suffix}'] +
            df[f'D{suffix}'] + df[f'N{suffix}']) / 10

df['TBM1'] = tinh_TBM(df, 1)
df['TBM2'] = tinh_TBM(df, 2)
df['TBM3'] = tinh_TBM(df, 6)
df
```

```python
#Cau 8
def xep_loai(tbm):
    if tbm < 5:
        return 'Y'
    elif tbm < 6.5:
        return 'TB'
    elif tbm < 8:
        return 'K'
    elif tbm < 9:
        return 'G'
    else:
        return 'XS'

df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)
df
```

```python
#Cau 9
def min_max_to_4(series):
    return 4 * (series - series.min()) / (series.max() - series.min())

df['US_TBM1'] = min_max_to_4(df['TBM1'])
df['US_TBM2'] = min_max_to_4(df['TBM2'])
df['US_TBM3'] = min_max_to_4(df['TBM3'])
df
```

```python
#Cau 10
def ket_qua_xt(row):
    if row['KT'] in ['A', 'A1']:
        diem = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    elif row['KT'] == 'B':
        diem = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    else:
        diem = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    return 1 if diem >= 5 else 0

df['KQXT'] = df.apply(ket_qua_xt, axis=1)
df
```

```python
#Cau 11
df.to_csv("processed_dulieuxettuyendaihoc.csv", index=False)
print("Đã lưu file processed_dulieuxettuyendaihoc.csv")
```
