```python
import pandas as pd
#Van de 1
column_names = [
    "Id", "Name", "Age", "Weight",
    "m0006", "m0612", "m1218",
    "f0006", "f0612", "f1218"
]
df = pd.read_csv(
    "patient_heart_rate.csv",
    names=column_names,
    engine="python",
    on_bad_lines="skip"
)
print(df.head())
```

```python
#Van de 2
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)

print(df.head())
```

```python
#Van de 3
weight = df['Weight']

for i in range(len(weight)):
    x = str(weight[i])

    if "lbs" in x:
        x = x.replace("lbs", "")
        value = float(x)
        kg = int(value / 2.2)
        df.at[i, 'Weight'] = str(kg) + "kgs"

print(df.head())
```

```python
#Van de 4
df.dropna(how="all", inplace=True)
print(df.head())
```

```python
#Van de 5
df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print(df.head())
```

```python
#Van de 6
df['Firstname'].replace({r'[^\x00-\x7F]+': ''}, regex=True, inplace=True)
df['Lastname'].replace({r'[^\x00-\x7F]+': ''}, regex=True, inplace=True)

print(df.head())
```

```python
#Van de 7 
import pandas as pd
import numpy as np

# a
print("Số lượng giá trị thiếu:")
print(df[['Age', 'Weight']].isnull().sum())
print("-" * 40)

# b
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# c
df = df.dropna(subset=['Age', 'Weight'], how='all')

# d
mean_age = df['Age'].mean()
print("Giá trị trung bình Age:", round(mean_age, 2))

# e
df['Age'] = df['Age'].fillna(mean_age)

print("\nDữ liệu sau khi xử lý missing values:")
print(df[['Age', 'Weight']])


```

```python
#Cau 11
df = pd.read_csv("patient_heart_rate.csv",engine="python",on_bad_lines="skip",header=None)
df.columns = ["id","fullname","age","weight","m0006","m0612","m1218","f0006","f0612","f1218"]
df[["Firstname", "Lastname"]] = df["fullname"].str.split(" ", n=1, expand=True)
df = df.drop(columns=["fullname"])

df = pd.melt(df,id_vars=["id", "age", "weight", "Firstname", "Lastname"],var_name="sex_and_time",value_name="PulseRate")
tmp = df["sex_and_time"].str.extract(r"([mf])(\d{2})(\d{2})")
tmp.columns = ["Sex", "Start", "End"]
df["Time"] = tmp["Start"] + "-" + tmp["End"]
df["Sex"] = tmp["Sex"]
df = df.drop(columns=["sex_and_time"])

df["PulseRate"] = df["PulseRate"].replace("-", np.nan)
df["PulseRate"] = pd.to_numeric(df["PulseRate"], errors="coerce")

time_order = {"00-06": 1, "06-12": 2, "12-18": 3}
df["TimeOrder"] = df["Time"].map(time_order)
df = df.sort_values(by=["id", "TimeOrder"])

def fill_pulse_rate(group):
    values = group["PulseRate"].copy()

    for i in range(len(values)):
        if pd.isna(values.iloc[i]):

            #trung binh lien truoc va lien sau
            prev_val = values.iloc[i-1] if i-1 >= 0 else np.nan
            next_val = values.iloc[i+1] if i+1 < len(values) else np.nan
            if not pd.isna(prev_val) and not pd.isna(next_val):
                values.iloc[i] = (prev_val + next_val) / 2
                continue

            #trung binh 2 gia tri lien truoc
            prev2 = values.iloc[max(0, i-2):i].dropna()
            if len(prev2) == 2:
                values.iloc[i] = prev2.mean()
                continue

            #trung binh 2 gia tri lien sau 
            next2 = values.iloc[i+1:i+3].dropna()
            if len(next2) == 2:
                values.iloc[i] = next2.mean()
                continue

            #trung binh cua cac gia tri huyet ap cua nguoi do
            person_mean = values.mean()
            if not pd.isna(person_mean):
                values.iloc[i] = person_mean

    group["PulseRate"] = values
    return group

df = df.groupby("id", group_keys=False).apply(fill_pulse_rate)

#trung binh cua nhom gioi tinh
df["PulseRate"] = df["PulseRate"].fillna(
    df.groupby("Sex")["PulseRate"].transform("mean")
)

#trung binh cua cac gia tri du lieu
overall_mean = df["PulseRate"].mean()
df["PulseRate"] = df["PulseRate"].fillna(overall_mean)

#muc on dinh y hoc
df["PulseRate"] = df["PulseRate"].fillna(70)

print("Số giá trị PulseRate còn thiếu:", df["PulseRate"].isna().sum())
print(df.head())
```
