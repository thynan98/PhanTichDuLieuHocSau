```python
#Phan 5
#Cau 1
from scipy import stats

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
df['T1'] = pd.to_numeric(df['T1'], errors='coerce')
t1 = df['T1'].dropna()
print("Số quan sát:", t1.count())
print("Mean:", t1.mean())
print("Median:", t1.median())
print("Std:", t1.std())
print("Variance:", t1.var())
print("Min:", t1.min())
print("Q1:", t1.quantile(0.25))
print("Q2 (Median):", t1.quantile(0.50))
print("Q3:", t1.quantile(0.75))
print("Max:", t1.max())
plt.figure(figsize=(6,4))
plt.boxplot(t1, vert=False)
plt.title("Box-Plot của biến T1")
plt.xlabel("Giá trị T1")
plt.show()
skewness = stats.skew(t1)
print("Skewness:", skewness)
plt.figure(figsize=(6,4))
plt.hist(t1, bins=10)
plt.title("Histogram phân phối biến T1")
plt.xlabel("Giá trị T1")
plt.ylabel("Tần số")
plt.show()
kurtosis = stats.kurtosis(t1)
print("Kurtosis:", kurtosis)
plt.figure(figsize=(5,5))
stats.probplot(t1, dist="norm", plot=plt)
plt.title("QQ-Plot của biến T1")
plt.show()

```

```python
#Cau 2
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
df['T1'] = pd.to_numeric(df['T1'], errors='coerce')

def phan_loai_t1(x):
    if pd.isna(x):
        return np.nan
    elif x < 5:
        return 'k'
    elif x < 7:
        return 'tb'
    elif x < 8:
        return 'kha'
    else:
        return 'g'
df['phanloait1'] = df['T1'].apply(phan_loai_t1)
df = df.dropna(subset=['T1', 'phanloait1'])
plt.figure(figsize=(8,5))
df.boxplot(column='T1', by='phanloait1')
plt.title("Box-Plot biến T1 theo nhóm phân lớp phanloait1")
plt.suptitle("")
plt.xlabel("Nhóm phân lớp T1")
plt.ylabel("Giá trị T1")
plt.show()
plt.figure(figsize=(10,6))
for nhom in df['phanloait1'].unique():
    data = df[df['phanloait1'] == nhom]['T1']
    plt.hist(data, bins=8, alpha=0.5, label=nhom)
plt.title("Histogram phân phối T1 theo nhóm phanloait1")
plt.xlabel("Giá trị T1")
plt.ylabel("Tần số")
plt.legend(title="phanloait1")
plt.show()
nhoms = df['phanloait1'].unique()
plt.figure(figsize=(12,8))
for i, nhom in enumerate(nhoms, 1):
    plt.subplot(2, 2, i)
    data = df[df['phanloait1'] == nhom]['T1']
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f"QQ-Plot T1 - Nhóm {nhom}")
plt.tight_layout()
plt.show()
```

```python
#Cau 3
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
df['T1'] = pd.to_numeric(df['T1'], errors='coerce')
df['DH1'] = pd.to_numeric(df['DH1'], errors='coerce')
df = df.dropna(subset=['T1', 'DH1'])
cov_T1_DH1 = df[['T1', 'DH1']].cov().iloc[0,1]
print("Covariance giữa T1 và DH1:", cov_T1_DH1)
corr_T1_DH1 = df[['T1', 'DH1']].corr().iloc[0,1]
print("Correlation giữa T1 và DH1:", corr_T1_DH1)
plt.figure(figsize=(7,5))
plt.scatter(df['T1'], df['DH1'], alpha=0.6)
plt.title("Biểu đồ Scatter khảo sát tương quan giữa T1 và DH1")
plt.xlabel("T1 (Biến độc lập)")
plt.ylabel("DH1 (Biến phụ thuộc)")
plt.grid(True)
plt.show()
```

```python
#Cau 4
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
df['T1'] = pd.to_numeric(df['T1'], errors='coerce')
df['DH1'] = pd.to_numeric(df['DH1'], errors='coerce')
df = df.dropna(subset=['T1', 'DH1', 'KV'])
print("Covariance và Correlation giữa T1 và DH1 theo từng khu vực:\n")
for kv in df['KV'].unique():
    df_kv = df[df['KV'] == kv]
    cov = df_kv[['T1', 'DH1']].cov().iloc[0,1]
    corr = df_kv[['T1', 'DH1']].corr().iloc[0,1]
    print(f"Khu vực {kv}:")
    print(f"  Covariance = {cov:.4f}")
    print(f"  Correlation = {corr:.4f}\n")
plt.figure(figsize=(10,6))
for kv in df['KV'].unique():
    df_kv = df[df['KV'] == kv]
    plt.scatter(df_kv['T1'], df_kv['DH1'], alpha=0.6, label=f"KV {kv}")
plt.title("Scatter khảo sát tương quan DH1 theo T1 trên từng khu vực")
plt.xlabel("T1")
plt.ylabel("DH1")
plt.legend(title="Khu vực")
plt.grid(True)
plt.show()
```
